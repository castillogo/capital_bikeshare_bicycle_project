### Read the data

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler
from sklearn.pipeline import make_pipeline
```

```python
StartStation31258
```

```python
result2=StartStation31258.groupby(["hour_local"]).size().to_frame('count').reset_index()
result3=StartStation31258.groupby(["hour_local"])
result3=result3.agg({'atemperatue':'mean','humidity':'mean','temperature':'mean', 'windSpeed':'mean'})
result3=result3.reset_index()
```

```python
result3
```

```python
result2
```

```python
result4=result3.merge(result2, on='hour_local')
```

```python
result4
```

```python
result4
#X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
#X.head()
```

```python
X = result4[['hour_local','atemperatue','humidity', 'temperature', 'windSpeed']]
y = result4['count']
```

### Split your data into training and test data and scale the data

```python
#y=y.set_index('count')
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
y_train
```

```python
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train))
X_train_scaled.head()
```

```python
X_test_scaled = pd.DataFrame(scaler.transform(X_test))
```

### Conduct PCA

First look at PCA with all features

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit(X_train_scaled, y_train)
```

```python
# Look at the eigenvector matrix
W = pd.DataFrame(pca.components_)
W.head()
#W.shape
```

```python
# Look at your newly created matrix
T = pd.DataFrame(pca.transform(X_train_scaled))
T.head()
#T.shape()
```

```python
import numpy as np
T_reconstructed = np.matmul(X_train_scaled, np.transpose(W))
```

```python
T_reconstructed.sort_index().head()
```

```python
# Plot the explained variance by all the features
pd.DataFrame(pca.explained_variance_ratio_).plot.bar()
```

```python
# Look at a cummulative plot
pd.Series(np.cumsum(pca.explained_variance_ratio_)).plot()
```

### Chose the first five principal components as they explain up to 90% of the variance

```python
pca5 = PCA(n_components=5)
pca5.fit(X_train_scaled)
```

```python
sum(pca5.explained_variance_ratio_)
```

```python
W5 = pd.DataFrame(pca5.components_, columns=['hour_local','atemperatue','humidity', 'temperature', 'windSpeed'])
```

```python
T5 = pd.DataFrame(pca5.transform(X_train))
```

```python
W5
```

```python
T5.shape
```

### Run a logisitc regression on all the original features

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=42)
```

```python
lr.fit(X_train_scaled, y_train)
lr.score(X_train_scaled, y_train), lr.score(X_test_scaled, y_test)
```

### Run a logistic regression on the 5 Principal Components

```python
pca3 = PCA(n_components=3)
pca3.fit(X_train_scaled)
```

```python
T_train = pd.DataFrame(pca3.transform(X_train_scaled))
T_test = pd.DataFrame(pca3.transform(X_test_scaled))
```

```python
lr_pca.score(T_train, y_train), lr_pca.score(T_test, y_test)
```
