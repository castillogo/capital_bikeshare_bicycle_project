```python
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
x = np.random.random(size=(100)) * 10
y = 4 * x + 3
```

```python
noise = np.random.normal(0.0, 4.0, size=(100))
ytrain = y + noise
```

```python
plt.plot(x, y, 'r-')
plt.plot(x, ytrain, 'bo')
```

```python
best_mse = np.Inf
best_a, best_b = None, None

for i in range(10000):
    a = np.random.random() * 10
    b = np.random.random() * 10
    yhat = a * x + b
    mse = np.sum((yhat - ytrain) ** 2) / yhat.shape[0]
    if mse <= best_mse:
        best_mse = mse
        best_a = a
        best_b = b
```

```python
yhat = best_a * x + best_b
```

```python
plt.plot(x, y, 'r-')
plt.plot(x, ytrain, 'bo')
plt.plot(x, yhat, 'b-')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
Xtrain = x.reshape(-1, 1)
m = LinearRegression(fit_intercept=False)
m.fit(Xtrain, ytrain)
m.score(Xtrain, ytrain) # this is a R^2, not an accuracy!!!
```

```python
ypred = m.predict(Xtrain)
```

```python
plt.plot(x, y, 'r-')
plt.plot(x, ytrain, 'bo')
plt.plot(x, ypred, 'b-')
```

```python
best_a, best_b
```

```python
m.coef_, m.intercept_
```

```python
mean_squared_error(ytrain, yhat)  # random
```

```python
mean_squared_error(ytrain, ypred) # scikit
```
