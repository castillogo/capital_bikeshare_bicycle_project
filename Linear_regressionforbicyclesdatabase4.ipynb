```python
df3 = df2[df2.workingday != 0]
df3 = df3[df3.season > 1]
df3 = df3[df3.season < 4]
df3
```

```python
x = df3[['temp']]
y = df3[['count']]
xneu=x.set_index('temp')
yneu=y.set_index('count')
```

```python
xarray = x.to_numpy()
yarray = y.to_numpy()
```

```python
best_mse = np.Inf
best_a, best_b = None, None
for i in range(1000):
    a = np.random.random()*10
    b = np.random.random()*10
    ycalc= a*xarray +b
    mse = np.sum((ycalc-yarray)**2)/ycalc.shape[0]
    if mse <= best_mse:
        best_mse= mse
        best_a = a
        best_b = b
```

```python
ycalc = best_a*x +best_b
```

```python
plt.plot(x, y, 'r-')
plt.plot(xarray, yarray, 'bo')
plt.plot(xarray,ycalc, 'g-')
```
